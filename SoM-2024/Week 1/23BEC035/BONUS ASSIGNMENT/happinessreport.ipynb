<h2>Task 2</h2>

<h5>Display the data of top 5 east asian countries based on 'Generosity'</h5>

```python
eastasiancountries = df1[df1["Regional indicator"] == "East Asia"]
eastasiancountries.sort_values(by="Generosity").head(5)
```

<h5>Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)</h5>

```python
plt.plot(eastasiancountries["Ladder score"], eastasiancountries["Generosity"], marker = 'o')
plt.title("Ladder score of the above 5 east asian countries based on generosity")
plt.xlabel("Ladder Score")
plt.ylabel("Generosity")
plt.show()
```

<h5>Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to show the relation these countries vs their 'Logged GDP per capita'</h5>

```python
countries = ['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']
df_2021 = df1[df1['Country name'].isin(countries)]
plt.scatter(df_2021["Logged GDP per capita"], df_2021["Country name"])
plt.title("Relation of countries vs their Logged GDP per capita")
plt.xlabel("Logged GDP per capita")
plt.ylabel("Countries")
plt.show()
```
